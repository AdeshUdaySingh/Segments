# Import

```python
import pandas as pd
import numpy as np
```

### Creating dataframe which will have factors

```python
df_factors = pd.read_excel('Segment - Factors.xlsx', sheet_name = 'Factors')
df_factors
```

### Creating dataframe which will have raw data

```python
df_data = pd.read_excel('Segment - Data.xlsx', sheet_name = 'Data')
df_data
```

### Creating empty dataframe which will have calculated values for each segment

```python
df_segment = df_data.copy()
df_segment
```

```python
for i in df_segment.columns :
    if i not in ['ID'] :
        df_segment.drop(i, axis = 1, inplace = True)
        
df_segment        
```

```python
for i in df_factors['Segment'].values :
    df_segment[i] = 0

df_segment    
```

### Creating a list to have all the column headers

```python
columns_list = []

for col in df_data.columns :
    if col not in ['ID'] :
        columns_list.append(col)
        
columns_list
```

### Calculating the values for each segment

```python
for i in range(len(df_data)) :
    for j in range(len(df_factors)) :
        result = 0
        for col in columns_list :
            result += (df_data.iloc[i][col] * df_factors.iloc[j][col]).sum()
        df_segment.iat[i, j+1] = result
```

```python
df_segment
```

```python
segment_only = []
for col in df_segment.columns :
    if col not in ['ID'] :
        segment_only.append(col)
        
segment_only        
```

### Creating a new column to have Max Score for each row

```python
df_segment['Max Score'] = df_segment[segment_only].max(axis = 1)
df_segment
```

### Creating new column to get segment name based on Max Score

```python
df_segment['Segment Name'] = df_segment[segment_only].idxmax(axis=1)
df_segment
```

### Creating new column to get the segment ID/code

```python
column_mapping = {col: i+1 for i, col in enumerate(df_segment.columns[1:])}  # +1 to match human-readable index (1-based)
df_segment['Segment Number'] = df_segment['Segment Name'].map(column_mapping)
df_segment
```

### Checking the segment counts

```python
df_segment_temp = df_segment.copy()
df_segment_temp
```

```python
drop_column = []

for col in df_segment_temp :
    if col not in ['Segment Name', 'Segment Number'] :
        drop_column.append(col)
        
drop_column        
```

```python
df_segment_temp.drop(columns = drop_column, axis = 1, inplace = True)
```

```python
df_segment_temp.groupby('Segment Name').count().reset_index()
```

### Exporting the output

```python
df_segment.to_excel('Segment Output.xlsx', index=False)
```
